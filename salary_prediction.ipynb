```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

**DATA LOADING & CLEANING**

```
!unzip archive.zip
```

```
df = pd.read_csv("Salary_Data.csv")
df
```

```
print(df.info())
print(df.isnull().sum())
```

```
df = df.dropna(subset=["Salary"]).copy()

# Impute missing values
num_columns = ["Age", "Years of Experience"]
for col in num_columns:
    df[col] = df[col].fillna(df[col].mean())

cat_columns = ['Gender', 'Education Level', 'Job Title']
for col in cat_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```
# Standardize the Education Level text
df['Education Level'] = df['Education Level'].replace({
    "Bachelor's Degree": "Bachelor's",
    "Master's Degree": "Master's",
    "phD": "PhD"
})
```

**MULTIPLE REGRESSION MODEL**

```
X = df.drop("Salary", axis=1)
y = df["Salary"]
X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```
# Train and Evaluate
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("--- MULTIPLE REGRESSION ---")
print("RMSE:", rmse)
print("R2 Score:", r2, "\n")
```

**SINGLE REGRESSION MODEL**

```
X_single = df[["Years of Experience"]]
```

```
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_single, y, test_size=0.2, random_state=42
)
```

```
# Train and Evaluate
single_model = LinearRegression()
single_model.fit(X_train_s, y_train_s)

y_pred_single = single_model.predict(X_test_s)
rmse_single = np.sqrt(mean_squared_error(y_test_s, y_pred_single))
r2_single = r2_score(y_test_s, y_pred_single)

print("--- SINGLE REGRESSION ---")
print("RMSE:", rmse_single)
print("R2 Score:", r2_single, "\n")
```

**COMPARE AND SAVE BEST MODEL**

```
if r2 > r2_single:
    best_model = model
    joblib.dump(best_model,'best_salary_model.pkl')
    print("Saved Multiple Regression model as 'best_salary_model.pkl'")
else:
    best_model = single_model
    joblib.dump(best_model,'best_salary_model.pkl')
    print("Saved Single Regression model as 'best_salary_model.pkl'")
```
